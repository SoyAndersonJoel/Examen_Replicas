```python
#arreglo de columnas porque me da error de datos en el power bi 
import re
def convertir_a_millones(valor):
    match = re.search(r'(\d+(\.\d+)?) million', valor)
    if match:
        numero = float(match.group(1))
        numero *= 1000000
        return int(numero)
    else:
        return None

dff['TCU'] = dff['TCU'].apply(convertir_a_millones)

dff
```
