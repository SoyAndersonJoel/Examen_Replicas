```python
import pandas as pd
from pymongo import MongoClient
```

```python
RankingJugadores = pd.read_csv("top5_leagues_player.csv")
RankingJugadores 
```

```python

#Comprobamos las dimensiones:
RankingJugadores.shape
```

```python

#Eliminamos los duplicados:
RankingJugadores = RankingJugadores.drop_duplicates()
#Comprobamos las dimensiones:
RankingJugadores.shape
```

```python
#Buscamos los valores nulos
RankingJugadores.isnull().sum()
```

```python
#los valores nulos de columnas numericas los remplazamos por la media
for columna in RankingJugadores.select_dtypes(include=['float64', 'int64']).columns:
        RankingJugadores[columna].fillna(RankingJugadores[columna].mean(), inplace=True)
```

```python
#Buscamos los valores nulos
RankingJugadores.isnull().sum()
```

```python
# Mostrar información general del DataFrame
print(RankingJugadores.info())
```

```python

# Mostrar las primeras filas del DataFrame
print(RankingJugadores.head())

```

```python


# Calcular el máximo y mínimo de la columna "Points"
max_puntos = RankingJugadores['price'].max()
min_puntos = RankingJugadores['price'].min()
print("\nMáximo del Precio:", max_puntos)
print("Mínimo de Precio:", min_puntos)
# Calcular el promedio de la columna "Points"
promedio_puntos = RankingJugadores['price'].mean()
print("Promedio de Precio:", promedio_puntos)
```

```python
# Convertir el DataFrame a un diccionario para facilitar su inserción en MongoDB
Insercion = RankingJugadores.to_dict(orient='records')
```

```python

# Conectar con la instancia de MongoDB 
client = MongoClient('localhost', 27017)  # Conexión local
db = client['FutbolData']  # Nombre de la base de datos
collection = db['RankingDeJugadores']  # Nombre de la colección
```

```python

# Insertar el diccionario en la colección de MongoDB
collection.insert_many(Insercion)
client.close()
```
