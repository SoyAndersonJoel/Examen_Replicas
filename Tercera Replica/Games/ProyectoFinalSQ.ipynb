```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

# Configurar el navegador
driver = webdriver.Chrome()  # o el navegador que prefieras

# Cargar la página inicial
driver.get("https://store.steampowered.com/search/?filter=topsellers&os=win")

# Simular desplazamiento hacia abajo para cargar más contenido
body = driver.find_element("tag name", "body")
for _ in range(100):  # Modifica según la cantidad de veces que necesites desplazar
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)  # Espera a que se cargue el contenido

# Obtener el HTML actualizado después del desplazamiento
html_content = driver.page_source

# Cerrar el navegador
driver.quit()

# Procesar el HTML con BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Ahora puedes encontrar y extraer los datos que necesitas

titulos=[]
precios=[]
windows=[]
macs=[]
linux=[]
fechas=[]
ratings=[]
porcentajes=[]
reseñas=[]
id=[]

games={}

# Ahora puedes encontrar y extraer los datos que necesitas
enlaces = soup.find_all("a", class_="search_result_row")
for pre in enlaces:
    precio = pre.find("div", class_="discount_final_price")
    if not precio:
        precios.append("0.00")
    else:
        precio1 = precio.text.strip()  # Obtener el texto del span y eliminar espacios en blanco adicionales
        precios.append(precio1)  # Agregar el título a la lista"""
        
for tit in enlaces:
    # Encontrar el span dentro de cada div
    span = tit.find("span", class_="title")
    if span:
        titulo = span.text.strip()  # Obtener el texto del span y eliminar espacios en blanco adicionales
        titulos.append(titulo)  # Agregar el título a la lista"""

for spans in enlaces:
    span = spans.find("span", class_="platform_img win")
    # Verificar si la clase 'win' está presente en el span
    if span:
        windows.append("True")  # Agregar True si la clase 'win' está presente
    else:
        windows.append("False")  # Agregar False en caso contrario
        
for mac in enlaces:
    span = mac.find("span", class_="platform_img mac")
    # Verificar si la clase 'win' está presente en el span
    if span:
        macs.append("True")  # Agregar True si la clase 'win' está presente
    else:
        macs.append("False")  # Agregar False en caso contrario  
        
for linu in enlaces:
    span = linu.find("span", class_="platform_img linux")
    # Verificar si la clase 'win' está presente en el span
    if span:
        linux.append("True")  # Agregar True si la clase 'win' está presente
    else:
        linux.append("False")  # Agregar False en caso contrario 

for fe in enlaces:
    div = fe.find("div", class_="col search_released responsive_secondrow")
    if div:
        fecha = div.text.strip()
        fechas.append(fecha)  # Agregar la fecha a la lista
        
for rat in enlaces:
    span_elements = rat.find("span", attrs={'data-tooltip-html': True})
    if not span_elements:
        ratings.append("none")
        reseñas.append("0")
        porcentajes.append("0")
    # Obtener el valor de data-tooltip-html
    else:
        tooltip_html = span_elements['data-tooltip-html']
        
        # Dividir el valor por "<br>" para obtener partes individuales
        tooltip_parts = tooltip_html.split('<br>')
        
        # La primera parte será "Variadas"
        rating_int = tooltip_parts[0]
        
        porcentaje_str=tooltip_parts[1]
        porcentaje_int = int(porcentaje_str.split()[1])
        
        reseñas_str=tooltip_parts[1].split()[5]
        reseñas_int=int(reseñas_str.replace(',',''))
        
        ratings.append(rating_int)
        reseñas.append(reseñas_int)
        porcentajes.append(porcentaje_int)
        
for enlace in enlaces:
    enlace_data = enlace["data-ds-appid"]
    id.append(enlace_data)

games['app_id']=id
games['title']=titulos
games['date_release']=fechas
games['win']=windows
games['mac']=macs
games['linux']=linux
games['rating']=ratings
games['positive_ratio']=porcentajes
games['user_reviews']=reseñas
games['price_final']=precios

print(len(id))
print(len(titulos))
print(len(fechas))
print(len(windows))
print(len(macs))
print(len(linux))
print(len(ratings))
print(len(porcentajes))
print(len(reseñas))
print(len(precios))

print(games)
```
